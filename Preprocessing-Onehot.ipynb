```python
import pandas as pd
import numpy as np

# Assuming df is your DataFrame and 'TotalCharges' is the column with the issue

# Step 1: Replace empty strings with NaN
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan)

# Step 2: Drop rows where 'TotalCharges' is NaN
df = df.dropna(subset=['TotalCharges'])

# Step 3: Convert the column to float
df['TotalCharges'] = df['TotalCharges'].astype(float)

# Now the DataFrame should no longer have any rows with empty or invalid 'TotalCharges' values
print(df['TotalCharges'].head())

```

```python
X=df.drop('Churn',axis=1)
Y=df['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,stratify=Y,random_state=21)
```

```python
numerical_attributes = x_train.select_dtypes(include=["int64", "float64"]).columns
categorical_attributes = x_train.select_dtypes(include="object").columns
```

```python
from sklearn.preprocessing import StandardScaler
Standard_Scaler = StandardScaler()
train_numerical_scaled=Standard_Scaler.fit_transform(x_train[numerical_attributes])
test_numerical_scaled=Standard_Scaler.transform(x_test[numerical_attributes])
```

```python
train_numerical_scaled = pd.DataFrame(train_numerical_scaled, columns=numerical_attributes, index=x_train.index)
test_numerical_scaled=pd.DataFrame(test_numerical_scaled, columns=numerical_attributes, index=x_test.index)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
# Encoding multiple columns. Unfortunately you cannot pass a list here so you need to copy-paste all printed categorical columns.
transformer = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), categorical_attributes))
```

```python
# Transforming
transformed = transformer.fit_transform(x_train)
# Transformating back
transformed_df = pd.DataFrame(transformed, columns=transformer.get_feature_names_out())
# One-hot encoding removed an index. Let's put it back:
transformed_df.index = x_train.index
# Joining tables
x_train = pd.concat([x_train, transformed_df], axis=1)
# Dropping old categorical columns
x_train.drop(categorical_attributes, axis=1, inplace=True)
# CHecking result
x_train.head()
```

```python
# Transforming
transformed = transformer.transform(x_test)
# Transformating back
transformed_df = pd.DataFrame(transformed, columns=transformer.get_feature_names_out())
# One-hot encoding removed an index. Let's put it back:
transformed_df.index = x_test.index
# Joining tables
x_test = pd.concat([x_test, transformed_df], axis=1)
# Dropping old categorical columns
x_test.drop(categorical_attributes, axis=1, inplace=True)
# CHecking result
x_test.head()
```

```python
# Concatenate the numerical and categorical DataFrames
x_train = pd.concat([train_numerical_scaled,x_train], axis=1)
x_test = pd.concat([test_numerical_scaled,x_test], axis=1)

# Check the result
print(x_train.shape)
print(x_test.shape)
x_test.head()
```
